```python
# define plot data

def plotData(X,y):
    
    #plot positive and negative indices
    pos = (y == 1).ravel()
    neg = (y == 0).ravel()

    fig, ax = plt.subplots(figsize = (6,4))

    plt.scatter(X[pos,0], X[pos,1], s=30, c='k', marker='+', linewidths=1)
    plt.scatter(X[neg,0], X[neg,1], s=30, c='y', marker='o', linewidths=1)
    
    return
```

```python
# plotting the svc hyperplane

def plot_svc(svc, X, y, h=0.02, pad=0.25):
    
    x_min, x_max = X[:, 0].min()-pad, X[:, 0].max()+pad
    y_min, y_max = X[:, 1].min()-pad, X[:, 1].max()+pad
    
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    
    Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plotData(X, y)
    plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.2)
    
    # Support vectors indicated in plot by vertical lines
    sv = svc.support_vectors_
    plt.scatter(sv[:,0], sv[:,1], c='r', marker='*', s=50, linewidths='1', label='Support Vectors')
    
    plt.legend()
    
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.show()
    
    print('Number of support vectors: ', svc.support_.size)    
    
    return
```

```python
# plot data for extracted X, y

plotData(X1,y1)
```

```python
# support vector machine from sklearn.svm

# with C = 1 (C is comparable to inverse of regularization parameter)

svc = svm.SVC(C=1.0, kernel='linear')
svc.fit(X1, y1.ravel())
plot_svc(svc, X1, y1)
```

```python
# support vector machine from sklearn.svm

# with C = 10 (C is comparable to inverse of regularization parameter)

svc = svm.SVC(C=10.0, kernel='linear')
svc.fit(X1, y1.ravel())
plot_svc(svc, X1, y1)
```

```python
# support vector machine from sklearn.svm

# with C = 100 (C is comparable to inverse of regularization parameter)

svc = svm.SVC(C=100.0, kernel='linear')
svc.fit(X1, y1.ravel())
plot_svc(svc, X1, y1)
```

```python
# defining gaussian (non-linear) kernel for the purpose of the exercise;

# (this is not needed for the SVC function of sklearn.svm; for that use kernel = 'rbf'
# see https://scikit-learn.org/stable/modules/svm.html)

def gaussianKernel(x1, x2, gamma):
    
    # gamma - bandwidth parameter
    
    norm = np.sum(np.square(x1-x2))
    
    similarity = (np.exp(-norm / (2 * (gamma ** 2))))
    
    return similarity
```

```python
# testing above function

x1 = np.array([1, 2, 1])
x2 = np.array([0, 4, -1])
gamma = 2

sim = gaussianKernel(x1, x2, gamma) # expected value should be about 0.324652

sim
```

```python
X3, y3, Xval, yval = data3['X'], data3['y'], data3['Xval'], data3['yval']

X3.shape, y3.shape, Xval.shape, yval.shape
```

```python
# plot data for new extracted X, y

plotData(X3,y3)
```

```python
# using poly / non-linear kernel ('poly') SVC

# with C = 1 (C is comparable to inverse of regularization parameter)

svc = svm.SVC(C=1.0, kernel='poly', degree=3, gamma=60)
svc.fit(X3, y3.ravel())
plot_svc(svc, X3, y3)
```

```python
def dataset3Params(X, y, Xval, yval):
    
    # list of C values and gamma values to try
    CValues     = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30]   
    gammaValues = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30]
         
    bestScore = 0
    bestParams = {'C': None, 'gamma': None}
    
    for c in CValues:        
        for g in gammaValues:                             
            svc = svm.SVC(C = c, kernel = 'rbf', gamma = g)
            svc.fit(X, y.ravel())
            score = svc.score(Xval, yval)        
            if score > bestScore:
                bestScore = score
                bestParams['C'] = c
                bestParams['gamma'] = g     
            
    return bestParams, bestScore
```

```python
bestParams, bestScore = dataset3Params(X3, y3, Xval, yval)
```

```python
bestScore, bestParams
```

```python
# use computed best params C, gamma to plot the hyper plane

c = 3
g = 30

svc = svm.SVC(C = c, kernel='poly', degree = 3, gamma = g)
svc.fit(X3, y3.ravel())
plot_svc(svc, X3, y3)
```
